### **2) Preparação e qualidade dos dados**

A análise abaixo mostrou que a coluna Cabin possuía 77,10% de dados ausentes, enquanto Age apresentava 19,87% e Embarked 0,22%. Por esse motivo, a coluna Cabin foi removida por conter muitos valores faltantes. Também foram excluídas as linhas que possuíam valores nulos em Age e Embarked. Após a limpeza, a base passou a ter 712 linhas e 11 colunas, sem valores ausentes nas variáveis restantes.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\lramos\OneDrive - Hexagon\Documents\FIAP\2026\Data Science\cp1_dados\Titanic-Dataset.csv")

print("=== HEAD() ===")
print(df.head())

print("\n=== INFO() ===")
df.info()

print("\n=== % DE NULOS POR COLUNA ===")
percentual_nulos = df.isnull().mean() * 100
print(percentual_nulos.sort_values(ascending=False))

df = df.drop_duplicates()
df = df.drop(columns=["Cabin"], errors="ignore")
df = df.dropna(subset=["Age", "Embarked"])

print("\n=== SHAPE APÓS LIMPEZA ===")
print(df.shape)

print("\n=== % DE NULOS APÓS LIMPEZA ===")
percentual_nulos_pos = df.isnull().mean() * 100
print(percentual_nulos_pos.sort_values(ascending=False))

print("\n=== HEAD()")
print(df.head())
```
